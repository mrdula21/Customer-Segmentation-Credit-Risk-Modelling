```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df = scaler.fit_transform(df)

print(df)
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
pca.fit(df)
```

```python
pca=pca.transform(df)
```

```python
pca.shape
```

```python
import sklearn.cluster as cluster
from sklearn.cluster import KMeans
%matplotlib inline
K=range(1,20)
wss=[]
for k in K:
    kmeans=cluster.KMeans(n_clusters=k,init="k-means++")
    kmeans=kmeans.fit(pca)
    wss.append(kmeans.inertia_)
```

```python
elbow=pd.DataFrame({'Clusters':K,'WSS':wss})
```

```python
plt.plot(K,wss, marker = 'o')
```

```python
kmeans_pca = KMeans(n_clusters=3,init='k-means++')
```

```python
label= kmeans_pca.fit_predict(pca)
```

```python
print(label)
```

```python
filtered_label0=pca[label==0]
```

```python
plt.scatter(filtered_label0[:,0] , filtered_label0[:,1])
plt.show()
```

```python
u_labels = np.unique(label)
for i in u_labels:
    plt.scatter(pca[label == i , 0] ,pca[label == i , 1] , label = i)
plt.legend()
plt.show()
```
